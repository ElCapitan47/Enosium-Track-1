```python
# you can run this file on google colab
# importing required libraries

import pandas as pd
import numpy as np
np.set_printoptions(precision=4)
import joblib 

from scipy import stats
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import precision_score, recall_score, accuracy_score 
```

```python
# reading train file
data = pd.read_csv("Track_1.csv")
```

```python
# dropping unnecessary columns
```

```python
data=data.drop(columns=['Unnamed: 0', 'Purpose of taking loan', '1=defaulted'])
```

```python
# pre processing the data
```

```python
for i in range(len(data['Age of the applicant in Number of Years'])):
  if data['Age of the applicant in Number of Years'][i] < 25:
    data['Age of the applicant in Number of Years'][i] = 1
  elif data['Age of the applicant in Number of Years'][i] < 35:
    data['Age of the applicant in Number of Years'][i] = 2
  elif data['Age of the applicant in Number of Years'][i] < 45:
    data['Age of the applicant in Number of Years'][i] = 3
  elif data['Age of the applicant in Number of Years'][i] < 55:
    data['Age of the applicant in Number of Years'][i] = 4
  elif data['Age of the applicant in Number of Years'][i] < 65:
    data['Age of the applicant in Number of Years'][i] = 5
  elif data['Age of the applicant in Number of Years'][i] < 75:
    data['Age of the applicant in Number of Years'][i] = 6
  else:
    data['Age of the applicant in Number of Years'][i] = 7
```

```python
print(data['Loan History'].unique())
data['Loan History'].replace({
    "no loans taken/all loans paid back duly": 5,
    "all loans at this bank paid back duly": 4,
    "existing loans paid back duly till now": 3,
    "delay in paying off loans in the past": 2,
    "critical account/other loans existing (not at this bank)": 1
}, inplace=True)
```

```python
print(data['amount in current account'].unique())
data['amount in current account'].replace({
    "less than 0": 0,
    "no current account": 1,
    "between 0 and 200": 2,
    "greater than 200": 3
}, inplace=True)
```

```python
print(data['amount in savings account'].unique())
data['amount in savings account'].replace({
    "no savings account": 0,
    "less than 100": 1,
    "between 100 and 500": 2,
    "between 500 and 1000": 3,
    "greater than 1000": 4
}, inplace=True)
```

```python
print(data['Owned property'].unique())
data['Owned property'].replace({
    "No property": 0,
    "car or other property": 1,
    "building society savings agreement/life insurance": 2,
    "Real Estate": 3
}, inplace=True)
```

```python
print(data['Type of job performed'].unique())
data['Type of job performed'].replace({
    "unskilled - resident": 0,
    "unemployed/ unskilled - non-resident": 1,
    "skilled employee / official": 2,
    "management/ self-employed/highly qualified employee/ officer": 3
}, inplace=True)
```

```python
cols=data[['Number of years of employment']]
for i in cols:
  data[i].fillna( data[i].mode().iloc[0] , inplace=True )
print(data['Number of years of employment'].unique())
data['Number of years of employment'].replace({
    "unemployed": 0,
    "less than a year": 1,
    "between 1 and 4 years": 2,
    "greater than 4 years": 3,
}, inplace=True)
```

```python
# converting data type object to int using labelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Fit the LabelEncoder object to the column
data['Guarantor or Debtor'] = le.fit_transform(data['Guarantor or Debtor'])
data['Marital Status'] = le.fit_transform(data['Marital Status'])
data['Other loans plans taken'] = le.fit_transform(data['Other loans plans taken'])
data['Working abroad or not'] = le.fit_transform(data['Working abroad or not'])
data['Is there telephone number available'] = le.fit_transform(data['Is there telephone number available'])
data['Type of Housing'] = le.fit_transform(data['Type of Housing'])
# data['Purpose of taking loan'] = le.fit_transform(data['Purpose of taking loan'])
```

```python
data.head(25)
```

```python
# dividing train set data into input and output data
# further this data is divided into train and test data
```

```python
X = data.drop("Loan Defaulted or not", axis=1) 
y = data["Loan Defaulted or not"] 

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=2) 
print(X_train.shape) 
print(y_train.shape)
print(X_test.shape) 
print(y_test.shape)
```

```python
# running XGBClassifier model
```

```python
from xgboost import XGBClassifier 

xgb = XGBClassifier(learning_rate =0.1,
 n_estimators=1000,
 max_depth=3,
 min_child_weight=1,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=1,
 seed=27)
xgb.fit(X_train, y_train) 

pred3 = xgb.predict(X_test)
print("Accuracy:", xgb.score(X_test, y_test))
```

```python
# reading the output file
```

```python
pred3 = ['%.3f' % elem for elem in list(pred3)]
submission = pd.DataFrame(list(zip(list(y), pred3)), columns =['Original Result', 'Predicted Result'])
submission.to_csv('UK4E60Z9.csv', index=False)
```
